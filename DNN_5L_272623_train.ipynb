# ONE_HOT 編碼轉換

```python
def one_hot_encode(labels):
    n_labels = len(labels)
    n_unique_labels = len(np.unique(labels))
    one_hot_encode = np.zeros((n_labels,n_unique_labels))
    one_hot_encode[np.arange(n_labels), labels] = 1

    return one_hot_encode
```

```python
def deal_data(train_path):
    y_train= list()
    X_train= list()
    f = open(train_path,'r')
    s = f.read()
    f.close()
    data_train = s.split("\n")
    data_train.pop()
    random.shuffle(data_train)
    #print(data_train[0])
    for i in data_train:
        data_t = i.split(" ",1)
        y_train.append(data_t[0])
        X_train.append(data_t[1])
        #print(data_t[0])
        #print(data_t[1])
    data_x_train = list()
    for i in range(0,len(X_train)):
        data_t2 = X_train[i].split(" ")
        data_t2.pop()
        #print(len(data_t2))
        data_x_train.append(np.zeros( len(data_t2) ))
        for i_2 in data_t2:
            temp = i_2.split(":")
            temp[0] = int(temp[0])-1
            temp[1] = float(temp[1])
            data_x_train[i][temp[0]] = temp[1]
            #print(type(temp[0]) )
            #print(type(temp[1]) )
    X_train = np.array(data_x_train)
    y_train_label = np.array(y_train)
        
    #X_test = np.array(data_x_test)
    #y_test_label = np.array(y_test)
    
    #print(X_train.shape)
    #print(y_train_label.shape)
    #print(X_test.shape)
    #print(y_test_label.shape)
    
    len_type = int( (len(data_x_train[0])/2) ** 0.5 ) 
    
    X_train = np.reshape(X_train, [-1,len_type,len_type,2])
    #X_test = np.reshape(X_test, [-1,2,len_type,len_type])
    #X_test = np.swapaxes(X_test,1,3)
    y_train_onehot = to_categorical(y_train_label)
    #y_test_onehot = to_categorical(y_test_label)
    print(X_train.shape)
    print(y_train_onehot.shape)
    #print(X_test.shape)
    #print(y_test_onehot.shape)
    return len_type,X_train,y_train_onehot
```

# Model 建立

```python
def model_lt2(len_type): 
    ops.reset_default_graph()
    keras.backend.clear_session()
    n_classes = 2  
    n_width = len_type
    n_height = len_type
    n_depth = 2
    n_inputs = n_height * n_width * n_depth  # total pixels

    model = Sequential()
    model.add(Dense(1, input_shape=(n_width,n_height,n_depth)))
        
    model.add(Flatten())
    #model.add(Dropout(rate = 0.25))
    
    model.add(Dense(units=1024, activation='relu'))
    #model.add(Dropout(rate = 0.25))
    
    model.add(Dense(units=512, activation='relu'))
    #model.add(Dropout(rate = 0.25))
        
    model.add(Dense(units=256, activation='relu'))
    #model.add(Dropout(rate = 0.25))
        
    model.add(Dense(units=128, activation='relu'))
    #model.add(Dropout(rate = 0.25))
        
    model.add(Dense(units=64, activation='relu'))
    #model.add(Dropout(rate = 0.25))
    
    model.add(Dense(units=32, activation='relu'))
    #model.add(Dropout(rate = 0.25))
        
    model.add(Dense(units=16, activation='relu'))
    #model.add(Dropout(rate = 0.25))
    
    model.add(Dense(units=8, activation='relu'))
    #model.add(Dropout(rate = 0.25))
        
    model.add(Dense(units=n_classes, activation='softmax'))
    #model.summary()

    return model
```

```python
def model_lt3(len_type):   
    ops.reset_default_graph()
    keras.backend.clear_session()
    
    n_classes = 2  
    n_width = len_type
    n_height = len_type
    n_depth = len_type
    n_inputs = n_height * n_width * n_depth * 2 # total pixels

    model = Sequential()
    model.add(Dense(1, input_shape=(n_width,n_height,n_depth,2)))
    
    
    model.add(Flatten())
    #model.add(Dropout(rate = 0.25))
    
    model.add(Dense(units=1024, activation='relu'))
    #model.add(Dropout(rate = 0.25))
    
    model.add(Dense(units=512, activation='relu'))
    #model.add(Dropout(rate = 0.25))
    
    model.add(Dense(units=256, activation='relu'))
    #model.add(Dropout(rate = 0.25))
    
    model.add(Dense(units=128, activation='relu'))
    #model.add(Dropout(rate = 0.25))
    
    model.add(Dense(units=64, activation='relu'))
    #model.add(Dropout(rate = 0.25))
    
    model.add(Dense(units=16, activation='relu'))
    #model.add(Dropout(rate = 0.25))
    
    model.add(Dense(units=n_classes, activation='softmax'))
    #model.summary()
        
    return model
```

```python
batch_size = 100
epochs = 10000
num_classes = 2
weight_decay = 1e-5
patience = [400]
for p in patience:
    for k in range(8,12):
        train_path,ver = get_data(k)
        len_type,X_train,y_train_onehot = deal_data(train_path)
        model_path = './model/DNN/'+ver+'_ND_p='+str(p)+'.hdf5'
        
        DNN_model = model_lt2(len_type)
        DNN_model.summary()
        DNN_model.compile(optimizer=Adam(lr=0.0001, beta_1=0.9, beta_2=0.999, epsilon=1e-08),loss = 'categorical_crossentropy',metrics=['accuracy'])
    
        es=EarlyStopping(
            monitor='val_accuracy',
            patience=p,
            verbose=0,
            restore_best_weights=True
        )
        mc=ModelCheckpoint(
            model_path,
            monitor='val_accuracy',
            verbose=1,
            save_best_only=True,
            save_weights_only=False,
            mode='max',
            period=1
        )
        
    
        history = DNN_model.fit(
            X_train,
            y_train_onehot,
            batch_size=batch_size,
            epochs = epochs,
            callbacks=[es,mc],
            validation_split = 0.1,
            verbose=1
        )
        
        
        with open(model_path+'_train_log.txt',"w") as f:
            f.write('Patience:' + str(p) + '\n')
            f.write('Train accuracy:' + str(max(history.history['accuracy']) ) + '\n')
            f.write('Val accuracy:' + str(max(history.history['val_accuracy']) ) + '\n')
        
    
```
